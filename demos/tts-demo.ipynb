```python
import json
from IPython.display import display, Audio
from huggingface_hub import InferenceClient
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
tts_client = InferenceClient("http://localhost:7777/models/tts")
```

```python
import wikipedia
page = wikipedia.page('Aritificial Intelligence')
```

```python
dir(page)
```

```python
ps = [e.strip() for e in page.content.split('\n\n\n')]
```

```python
summarized_ps = [pipe(p) for p in ps[:10]]
```

```python
len(summarized_ps)
```

```python
res = tts_client.text_to_speech(summarized_ps[0][0]['summary_text'])
res = json.loads(res)
display(Audio(data=res, rate = 16000, autoplay=True))
```
